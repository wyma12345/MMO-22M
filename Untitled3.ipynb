# Возмем датасет "Аварии в вымышленных годородах"

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('traffic.csv')

# Make default density plot
sns.kdeplot(data = df, x='Speed', hue="Weather")
```

На графике видно, что большинство аварий происходит в снежную погоду на скоростях примерно в 60 км/ч (37 миль/час)

```python
# Make default density plot
sns.kdeplot(data = df, x='Speed')
```

Наибольшее кол-во аварий просисходит на скорости в 60-62 км/ч

```python
fig, ax = plt.subplots()
ax.set_xlim(0,1)
ax.set_ylim(0,24)
sns.kdeplot(data = df, x='Traffic Density', y = "Hour Of Day", fill=True, ax=ax)
```

Аварии чаще всего проиходят в промежутке между 6 и 22 часами(Когда наблюдается самый большой трафик) и при невысокой плотности движения(чем выше плотность, тем ниже скорость, а значит ниже шанс попасть в аварию)

```python
ax.set_xlim(0,1)
sns.histplot(data=df, x="Is Peak Hour")
```

Больниство аварий происходит не в час пик

```python
param = df["Is Peak Hour"].value_counts()
pie, ax = plt.subplots(figsize=[10,6])
plt.pie(param, labels=["Is Peak Hour", "Is not Peak Hour"], pctdistance=0.5, autopct='%.0f%%')
plt.show()


# Calculate the frequencies of each value

# df2 = pd.DataFrame({
#     "Pick": df["Is Peak Hour"].value_counts(),
#     "NoPick": df["Is Peak Hour"].value_counts()
# })
# df2
```

```python
sns.histplot(data=df, x="Economic Condition", hue="Random Event Occurred", multiple="stack")
```

В городах, находящихся в стабильной фазе развития, влияние случайных факторов на происшествия ниже,скорее всего, это связано с тем, что в бурноразвивающихся городах идет постояная перестройка дорог, что вводит путнаицу в дорожное движение. В городах находящихся в процесе стогнации у города нехватает средств на улучшение дорожного движения и починку дорожного полотна.

# Анализ датасета воторого "Аварии"

```python
df = pd.read_csv('car_crash.csv')

# чаще попадают в аварии 20 летние
sns.kdeplot(data = df, x='ageOFocc')
```

Чаще всего в аварии попадают 19-20 летние, после чего возраст резко идет вниз. (Чем моложе человек, тем больше в нем задора)

```python
# в машинах каких годов чаще умирали
fig, ax = plt.subplots()
ax.set_xlim(1970,2005)
sns.histplot(data=df, x="yearVeh",  hue="dead", ax=ax)
```

Самые опасные автомобили, были автомобили, произведеные с 1985 до 1997 года. Чаще всего аварии происходили на автомобилях, выпущенных в 1995 году. Однако, хоть количество аварий на автомобилях, выпщенных после 1995 года падает не сильно, смертность идет в низ, это связано с улучшение безопасности выпускаемых автомобилей.

```python
# в аварии чаще попавдают мужчины
param = df["sex"].value_counts()
pie, ax = plt.subplots(figsize=[10,6])
plt.pie(param, labels=["man", "woman"], pctdistance=0.5, autopct='%.0f%%')
plt.show()
```

В авариях чаще погибают мужчины

```python
#умирает больше мужчин чем женщин
param = df["sex"].value_counts()
param = df.groupby('sex')['dead'].value_counts()

pie, ax = plt.subplots(figsize=[10,6])
plt.pie([param[1], param[3]], labels=["woman dead", "man dead"], pctdistance=0.5, autopct='%.0f%%')
plt.show()
```

В авариях чаще умирают мужчины

(процент умерших мужчин выше, чем процент мужчин попавших в аварию)

```python
# сколько выживших было пристегнуто и с подушками
param = df.loc[df["dead"] == "alive"]
param = param.groupby('airbag')['seatbelt'].value_counts()
plt.pie(param, labels=["airbag and belted", "airbag and NOT belted", "NOT airbag and belted", "NOT airbag and NOT belted"], pctdistance=0.5, autopct='%.0f%%')
plt.show()
param
```

Большинство выживших было пристегнуто и в их машинах имелись подушки безопасности

```python
# сколько умерших было пристегнуто и с подушками
param = df.loc[df["dead"] == "dead"]
param = param.groupby('airbag')['seatbelt'].value_counts()
plt.pie(param, labels=["airbag and NOT belted", "airbag and belted", "NOT airbag and NOT belted", "NOT airbag and belted"], pctdistance=0.5, autopct='%.1f%%')
plt.show()
param
```

Чаще умирали люди, которые не были пристегнуты и в их машинах не было подушек безопасности

```python
# сколько умерших было пристегнуто и с подушками при фронтовом ударе
param = df.loc[(df["dead"] == "dead") & (df["frontal"] == 1)]
param = param.groupby('airbag')['seatbelt'].value_counts()
plt.pie(param, labels=["airbag and NOT belted", "airbag and belted", "NOT airbag and NOT belted", "NOT airbag and belted"], pctdistance=0.5, autopct='%.1f%%')
plt.show()
param
```
